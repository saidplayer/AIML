```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import normalize
from sklearn.metrics import r2_score, root_mean_squared_error as rmse
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# def main():
data = make_regression(n_samples=200, n_features=10, n_informative=6, n_targets=1, noise=10, random_state=42)
x = data[0].reshape((-1,10))
y = data[1].reshape((-1,1))
scaler = normalize
x = scaler(x)
# y = scaler(y)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)

model_lin   = LinearRegression()
model_ridge = Ridge(alpha = 0.1)
model_lasso = Lasso(alpha = 0.5)

model_lin.fit(x_train, y_train)
print(r2_score(y_test, model_lin.predict(x_test)))
print(rmse(y_test, model_lin.predict(x_test)))
print(model_lin.coef_)

model_ridge.fit(x_train, y_train)
print(r2_score(y_test, model_ridge.predict(x_test)))
print(rmse(y_test, model_ridge.predict(x_test)))
print(model_ridge.coef_)


model_lasso.fit(x_train, y_train)
print(r2_score(y_test, model_lasso.predict(x_test)))
print(rmse(y_test, model_lasso.predict(x_test)))
print(model_lasso.coef_)

plt.figure(figsize=(12, 4))
plt.plot(x_test.T[0], y_test, 'ok')
plt.plot(x_test.T[0], model_ridge.predict(x_test), 'xb')
plt.plot(x_test.T[0], model_lin.predict(x_test), 'xr')
plt.plot(x_test.T[0], model_lasso.predict(x_test), 'xg')


# if __name__ == "__main__": 
#     main()
```
